```python
import os
from dotenv import load_dotenv
import json
import requests
import re
load_dotenv() 
```

```python
token = os.getenv("token")
```

```python
headers = {"Authorization": f"token {token}"}
```

```python
url = "https://api.github.com/repos/ironhack-datalabs/datamad1020-rev/forks"
```

```python
res = requests.get(url, headers=headers)
```

```python
print(res)
```

```python
eljeison = res.json()
```

```python
print(eljeison)
```

```python
forks = []
for i in eljeison:
    for x in i.keys():
        if (x=="language") and i.get("language") not in forks:
            forks.append(i.get("language"))
print(forks)
```
